```python
# importing the libraries
import os
import pandas as pd
import numpy as np
import sys
import glob
from datetime import datetime, timedelta
```

```python
# list of necessary columns from flight data
l_cols = ['YEAR','QUARTER','MONTH','DAY_OF_MONTH','DAY_OF_WEEK','FL_DATE','TAIL_NUM',
          'ORIGIN','ORIGIN_CITY_NAME','DEST','DEST_CITY_NAME','DEP_TIME','DEP_DELAY',
          'DEP_DELAY_NEW','DEP_DEL15','CANCELLED']
```

```python
# combining all the flight data csvs together

path = "/content/drive/MyDrive/data/filght_data/"

filenames = glob.glob(path + "/*.csv")
month_data = []

for file in filenames:
  print(file)
  data = pd.read_csv(file)[l_cols]
  month_data.append(data)
```

```python
# concatanating all the files

flight_data = pd.concat(month_data,axis=0)
```

```python
del month_data
```

```python
flight_data.head()
```

```python
flight_data.isna().sum()
```

```python
# dropping rows where the filght tail number or the departure time is missing
flight_data.dropna(subset=['TAIL_NUM','DEP_TIME','DEP_DELAY'], inplace=True)
```

```python
flight_data.isna().sum()
```

```python
# checking the data types of the columns
flight_data.dtypes
```

```python
# correcting the data types of the columns as needed
flight_data['FL_DATE'] = pd.to_datetime(flight_data['FL_DATE'])
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].astype(int).astype(str)
```

```python
flight_data.dtypes
```

```python
# flight_data.to_csv('/content/drive/MyDrive/data/flight_data_combined.csv')
```

```python
# left padding the DEP_TIME so that all entries have HHMM format
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].str.pad(4,side='left',fillchar='0')
```

```python
flight_data.head()
```

```python
# extracting the departure hour and the minutes from DEP_TIME column
flight_data['DEP_HOUR'] = flight_data['DEP_TIME'].str[0:2]
flight_data['DEP_MIN'] = flight_data['DEP_TIME'].str[-2:]
```

```python
flight_data.head()
```

```python
# we have some row where dep hour is 24, need to make the date into next day and hour to 00
flight_data[flight_data.DEP_HOUR == "24"].head()
```

```python
flight_data['FL_DATE'] = np.where(
    flight_data['DEP_HOUR'] == "24", flight_data['FL_DATE'] + timedelta(days=1),
    flight_data['FL_DATE']
)

flight_data['DEP_HOUR'] = np.where(
    flight_data['DEP_HOUR'] == "24", "00",
    flight_data['DEP_HOUR']
)
```

```python
# reading aircraft data and dropping the na values
aircraft_data = (
    pd.read_csv('/content/drive/MyDrive/data/aircraft_data/tail_n_aircraft.csv')[['TAIL_NUM','Aircraft type']]
    .dropna(subset=['Aircraft type'])
    .drop_duplicates()
    )
```

```python
aircraft_data.head()
```

```python
# joining the flight data and the aircraft type dataset
combined_aircraft_type = (
    flight_data
    .merge(
       aircraft_data,
       on=['TAIL_NUM'],
       how='left'
    )
    .dropna(subset=['Aircraft type'])
)
```

```python
del flight_data
del aircraft_data
```

```python
combined_aircraft_type.head()
```

```python
# combined_aircraft_type.to_csv('/content/drive/MyDrive/data/aircraft_data_combined.csv')
```

```python
# reading weather data
weather_data = (
    pd.read_csv('/content/drive/MyDrive/data/weather_data/atlanta_weather.csv', parse_dates=['FL_DATE'])
    .drop(['Unnamed: 0', 'snow', 'wpgt', 'tsun'],axis=1)
    )
```

```python
weather_data.head()
```

```python
# filling the DP_HOUR columne to have 2 digits
weather_data['DEP_HOUR'] = weather_data['DEP_HOUR'].astype('str').str.pad(2,side='left',fillchar='0')
```

```python
weather_data.dtypes
```

```python
combined_aircraft_type.head()
```

```python
# joining the weather data to create the final dataset

final_data = (
    combined_aircraft_type
    .merge(
        weather_data,
        on=['FL_DATE','DEP_HOUR'],
        how='left'
    )
)
```

```python
del combined_aircraft_type
del weather_data
```
